```python
df["body_mass_g"]=df["body_mass_g"].fillna(df["body_mass_g"].median())
df["bill_length_mm"]=df["bill_length_mm"].fillna(df["bill_length_mm"].median())
df["bill_depth_mm"]=df["bill_depth_mm"].fillna(df["bill_depth_mm"].median())
df["flipper_length_mm"]=df["flipper_length_mm"].fillna(df["flipper_length_mm"].median())
df["sex"]=df["sex"].fillna(df["sex"].mode()[0])
# now filling the null values
# numeric values are filled with median beacuse the graph is right skewed so from this it little bit comes to 
# shape
# for catogorical columns we fill it with mode 0f [0] which means from man we fill it
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
# now importing the libliries from sklearn for forther opertion and also to encode the catogorial columns to numeric
```

```python
catagoric=["species","island","sex"]
numeric=["bill_length_mm","bill_depth_mm","flipper_length_mm"]
# now divide this is two parts for StandardScaler and OneHotEncoder
```

```python
df.isnull().sum()
```

```python
prepro=ColumnTransformer(transformers=[("num",StandardScaler(),numeric),
                        ("cata",OneHotEncoder(drop="first",sparse_output=False),catagoric)])
# now the stantdard and hot encoder are doing here this is the logic behind it
```

```python
# sns.pairplot(df)
```

```python
x=df.drop("body_mass_g",axis=1)
y=df["body_mass_g"]
# we will find the mass_body from the other feathers
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
# divideing the data into training and test 
# where traning data is 80% and testing data is 20%
```

```python
x_train_p=prepro.fit_transform(x_train)
x_test_p=prepro.transform(x_test)
# now replacing th variable and add the standard scaler and one hot encoder for forther operation
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train_p,y_train)
```

```python
pred=model.predict(x_test_p)
# predict the values here
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,pred)
# to find the accuracy here 
# which i 84%
```

```python
from sklearn.tree import DecisionTreeRegressor,plot_tree
# applying tree algorithem here
```

```python
model2=DecisionTreeRegressor()
model2.fit(x_train_p,y_train)
# train the model 
```

```python
pred2=model2.predict(x_test_p)
pred2
# predict the values here
```

```python
r2_score(y_test,pred2)
# poor r2_score
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Assuming x_test_p, y_test, pred (LinearRegression) and pred2 (DecisionTreeRegressor) are already computed
# If not, re-run the preprocessing and predictions as shown in the notebook.

# 1. Compute additional metrics
metrics = {
    'Model': ['Linear Regression', 'Decision Tree Regressor'],
    'R²': [r2_score(y_test, pred), r2_score(y_test, pred2)],
    'MAE': [mean_absolute_error(y_test, pred), mean_absolute_error(y_test, pred2)],
    'MSE': [mean_squared_error(y_test, pred), mean_squared_error(y_test, pred2)],
    'RMSE': [np.sqrt(mean_squared_error(y_test, pred)), np.sqrt(mean_squared_error(y_test, pred2))]
}

metrics_df = pd.DataFrame(metrics)
print(metrics_df)

# 2. Visual comparison
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Model Comparison: Linear Regression vs Decision Tree', fontsize=14)

# Actual vs Predicted - Linear Regression
axes[0,0].scatter(y_test, pred, alpha=0.6, edgecolor='k')
axes[0,0].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
axes[0,0].set_xlabel('Actual Body Mass (g)')
axes[0,0].set_ylabel('Predicted Body Mass (g)')
axes[0,0].set_title('Linear Regression')

# Actual vs Predicted - Decision Tree
axes[0,1].scatter(y_test, pred2, alpha=0.6, edgecolor='k', color='green')
axes[0,1].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
axes[0,1].set_xlabel('Actual Body Mass (g)')
axes[0,1].set_ylabel('Predicted Body Mass (g)')
axes[0,1].set_title('Decision Tree Regressor')

# Residuals - Linear Regression
residuals_lr = y_test - pred
axes[1,0].scatter(pred, residuals_lr, alpha=0.6, edgecolor='k')
axes[1,0].axhline(y=0, color='r', linestyle='--')
axes[1,0].set_xlabel('Predicted Body Mass (g)')
axes[1,0].set_ylabel('Residuals')
axes[1,0].set_title('Residuals (Linear Regression)')

# Residuals - Decision Tree
residuals_dt = y_test - pred2
axes[1,1].scatter(pred2, residuals_dt, alpha=0.6, edgecolor='k', color='green')
axes[1,1].axhline(y=0, color='r', linestyle='--')
axes[1,1].set_xlabel('Predicted Body Mass (g)')
axes[1,1].set_ylabel('Residuals')
axes[1,1].set_title('Residuals (Decision Tree)')

plt.tight_layout()
plt.show()

# 3. Bar chart of R² and RMSE
fig, ax = plt.subplots(figsize=(8,5))
bar_width = 0.35
index = np.arange(2)

r2_scores = [metrics_df['R²'][0], metrics_df['R²'][1]]
rmse_scores = [metrics_df['RMSE'][0], metrics_df['RMSE'][1]]

bars1 = ax.bar(index, r2_scores, bar_width, label='R²', color='blue')
bars2 = ax.bar(index + bar_width, rmse_scores, bar_width, label='RMSE (g)', color='orange')

ax.set_xlabel('Model')
ax.set_ylabel('Score')
ax.set_title('Performance Comparison')
ax.set_xticks(index + bar_width/2)
ax.set_xticklabels(['Linear Regression', 'Decision Tree'])
ax.legend()

# Add value labels on bars
for bar in bars1:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.01, f'{height:.3f}', ha='center', va='bottom')
for bar in bars2:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 5, f'{int(height)}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
